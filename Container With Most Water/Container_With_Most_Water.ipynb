# Container With Most Water
You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are **(i, 0) and (i, height[ i ])**.

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.



    Example 1:
    Input: height = [1,8,6,2,5,4,8,3,7]
    Output: 49
    Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7].
    In this case, the max area of water (blue section) the container can contain is 49.

    Example 2:
    Input: height = [1,1]
    Output: 1

```python
# First lets understand the question a little better as I struggled with the explanation of the question too on LeetCode.
# To better understand this question lets visualize it
import numpy as np
import matplotlib.pyplot as plt

# for the x-axis and the y-axis, lets say
# for the x and the y values this depends on the user's input so lets take that first
n = int(input('Enter the width of the container: '))
height = []
for i in range(n):
  num = int(input('Enter the height of the container at this instance:  '))
  height.append(num)

print('The container heights =', height)
```

```python
# Now lets visualize the entered input from the user to better understand the question
x_values = [i for i in range(n)]
y_values = height.copy()

print("To Better understand the bar shown below, let's even map the x value to its corresponding y value")
print("Shown (x, y) as (Width, Height)")
for i in range(n):
  print(f'({x_values[i]}, {y_values[i]})', end = ', ')

fig = plt.figure(figsize = (8, 5))

# plt.plot([min(height[x], height[y]) for i in range(y - x + 1)], [i for i in range(x, y+1)], 'r--') we'll implement after we find the answer to our problem
plt.bar(x_values, y_values, color = 'black', width = 0.3)
plt.xlabel("Width of the container")
plt.ylabel("Height of the container")
plt.title("Container visualizing its height and width")
plt.show()
```

Now the problem asked is we have to match the width and height to make a container that can hold the **maximum** amount of water when poured in it.
Which is to mean to calculate the maximum area given the heights of the container.

```python
class Solution:
  def maxArea(self, height):

    left = 0
    right = n - 1
    maxArea = 0
    global x
    global y

    while left < right:
      currentArea = min(height[left], height[right]) * (right - left) # we take the minimum height as it is the height between two bars in a container that can hold water,
                                                                      # Example, if a container has one side of 1m and another of 2m, water will slip from that 1m side.
      maxArea = max(maxArea, currentArea) #we need to find the max of all area so we use two pointer from left and right to solve the problem



      if height[left] < height[right]:
        left += 1
      else:
        right -= 1

      if currentArea == maxArea:
        x = left
        y = right + 1
        print(f'({x}, {y})')

    return maxArea
```

```python
output = Solution()
print('The maximum area will be ')
print(output.maxArea(height))
```
