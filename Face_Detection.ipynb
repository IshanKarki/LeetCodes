```python
import cv2
imagePath = '/content/drive/MyDrive/F1Soft.jpg'

#read the image
img = cv2.imread(imagePath)
#image shape
print(img.shape)
plt.imshow(img)
```

```python
#lets convert the colored image into grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray_image.shape)
plt.imshow(gray_image)
```

```python
#loading the classifier
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

#perform the face detection
face = face_classifier.detectMultiScale(gray_image, scaleFactor = 1.1, minNeighbors = 20, minSize = (20, 20))

#drawing the rectangular bound box
for (x, y, w, h) in face:
  cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

```python
#displaying the image
#converting the image from bgr to rgb
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#using matplotlib to display the image
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.imshow(img)
plt.imshow(img_rgb)
plt.axis('off')
```
