# Find the Difference
You are given two strings ***s*** and ***t***.

String ***t*** is generated by random shuffling string ***s*** and then add one more letter at a random position.

Return the letter that was added to ***t***.



    Example 1:
    Input: s = "abcd", t = "abcde"
    Output: "e"
    Explanation: 'e' is the letter that was added.

    Example 2:
    Input: s = "", t = "y"
    Output: "y"


```python
#lets take input s from the user and lets add one more letter at a random position
from numpy import random

s = input("Enter the random string:  ")
t = s + random.choice(['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z'])
print(t)

t = list(t)
random.shuffle(t)
t = ''.join(t)

print(t)
```

```python
class Solution:
  def findTheDifference(self, s, t):

    # we loop through entire sequence of strings in t
    # and check if the particular string element in t is not present in s but this alone is not enough as in
    # cases like s = 'a', t = 'aa', both iterables in t, that is a is present in s, thus to solve such issues we count,
    # we check if the count('a') in t > count('a') in s which is 2 > 1 thus we return a, also
    # say s = 'abcd', t = 'abcde', then we at 'e' in t, count('e') in t = 1 > count('e') in s = 0, 1>0 thus, we return e.

    for i in t:
      if i not in s or t.count(i) > s.count(i):
        return i
```

```python
output = Solution()
print(f'The random string is {s}')
print(f'The shuffled string in t is {t}')
print('Thus, the random letter that was added to t is,')
print(output.findTheDifference(s, t))
```
